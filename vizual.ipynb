```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.ticker as mticker

plt.style.use('seaborn-v0_8-whitegrid')
df = pd.read_csv('search_results.csv')

sizes = df['Size']
time_columns = ['Linear', 'BST', 'RBT', 'Hash', 'Multimap']
custom_time_columns = ['Linear', 'BST', 'RBT', 'Hash']

plt.figure(figsize=(10, 6))
for col in custom_time_columns:
    plt.plot(sizes, df[col], marker='o', linestyle='-', label=col)
plt.title('   ( )')
plt.xlabel(' ')
plt.ylabel('  ()')
plt.legend()
plt.grid(True, which="both", ls="--", alpha=0.7)
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 6))
for col in custom_time_columns:
    plt.plot(sizes, df[col], marker='o', linestyle='-', label=col)
plt.yscale('log')
plt.title('   (  Y)')
plt.xlabel(' ')
plt.ylabel('  () (log )')
plt.legend()
plt.grid(True, which="both", ls="--", alpha=0.7)
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 6))
for col in time_columns:
    plt.plot(sizes, df[col], marker='o', linestyle='-', label=col)
plt.title('   ( )')
plt.xlabel(' ')
plt.ylabel('  ()')
plt.legend()
plt.grid(True, which="both", ls="--", alpha=0.7)
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 6))
for col in time_columns:
    plt.plot(sizes, df[col], marker='o', linestyle='-', label=col)
plt.yscale('log')
plt.title('   (  Y)')
plt.xlabel(' ')
plt.ylabel('  () (log )')
plt.legend()
plt.grid(True, which="both", ls="--", alpha=0.7)
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 6))
plt.plot(sizes, df['Collisions'], marker='o', linestyle='-', color='teal')
plt.title('     ')
plt.xlabel(' ')
plt.ylabel(' ')
formatter = mticker.FuncFormatter(lambda x, p: format(int(x), ','))
plt.gca().yaxis.set_major_formatter(formatter)
plt.grid(True, which="both", ls="--", alpha=0.7)
plt.tight_layout()
plt.show()
```
